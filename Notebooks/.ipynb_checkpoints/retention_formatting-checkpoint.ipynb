```python
import pandas as pd
```

```python
general_results = pd.read_csv('Data/general_results.csv')
```

```python
general_results = general_results[general_results.audiencecohort != 'holdoutgroup_braze']
```

```python
def audience_cohort_parsing(row):
    if row == 'paiddis':
        return 'Paid Display'
    elif row == 'paidsoc':
        return 'Paid Social'
    else:
        return 'All Paid Media'

def targetable_bool(row):
    if row == 'targetablegroup':
        return 'Targetable'
    else:
        return 'Holdout'

def main_kpi(input_df):
    input_df = input_df.merge(input_df.audiencecohort.str.split("_", expand = True), left_index = True, right_index = True)
    input_df['Channel'] = input_df[1].apply(lambda row: audience_cohort_parsing(row))
    input_df['Targetable_Boolean'] = input_df[0].apply(lambda row: targetable_bool(row))
    output_df = input_df[['Channel', 'Targetable_Boolean', 'cohort_size', 'conversions_watched_video', 'conversion_rate', 'repertoire', 'avg_starts', 'usage', 'avg_completion_rate', 'all_starts', 'all_completes']]
    column_mapping = {
        'cohort_size' : 'Cohort Size',
        'conversions_watched_video': 'Total User Return',
        'conversion_rate': 'Return Rate',
        'repertoire': 'Repertoire', 
        'avg_starts': 'Average Starts',
        'usage': 'Usage (Hours)',
        'avg_completion_rate': 'Average Completion Rate',
        'all_starts': 'All Starts',
        'all_completes': 'All Completes'
    }
    output_df = output_df.rename(columns = column_mapping)[['Channel', 'Targetable_Boolean', 'Cohort Size', 'Return Rate', 'Total User Return', 'Repertoire', 'Usage (Hours)']].sort_values(['Channel', 'Targetable_Boolean']).reset_index(drop = True)
    return output_df
```

```python
main_kpi(general_results)
```
